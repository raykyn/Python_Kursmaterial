## Mehrere Zahlen aufs Mal: Listen
Oft reicht es aber nicht, eine einzelne Zahl in einer einzelnen Variable abzuspeichern. Sehen wir uns einen neuen Datentypen an, die *Liste* (list). Listen ermöglichen uns, andere Objekte, zum Beispiel Zahlen, in einer Sequenz zu erfassen und bei Bedarf abzurufen.

```python
# Eine Liste wird durch eckige Klammern markiert
noten = [4, 5, 5, 4.5, 3.5, 5.5, 4]
```

Listen sind genau so Objekte wie es auch integer und floats sind. Wir können sie also zum Beispiel als Parameter verwenden:

```python
noten_summe = sum(noten)

print(noten_summe)
```

Einzelne Elemente können wir uns folgendermassen ansehen:

```python
print(noten[3])
```

Sortierte sequentielle Objekte (nicht nur Listen) speichern ihren Inhalt unter sogenannten *Indizes*. Wir können also einen integer als Index eingeben, um ein Element an einer bestimmten Position in der Liste abzurufen. Umgekehrt können wir auch den Index abrufen von einem Element in der Liste.

```python
print(noten.index(3.5))
```

Beachte, dass der Index immer bei 0 beginnt! Wenn eine Liste also zum Beispiel 8 Elemente enthält, dann hat das letzte Element den Index 7. Wir können den Index übrigens auch rückwärts abrufen, so:

```python
print(noten[-2])
```

Es gibt einen Haufen [Funktionen, mit denen wir Listen manipulieren können](https://www.w3schools.com/python/python_ref_list.asp):

```python
# Zum Beispiel können wir Objekte hinzufügen
noten.append(6)

# oder Objekte entfernen
noten.remove(3.5)

# alle Elemente umdrehen
noten.reverse()

# oder die Elemente sortieren
noten.sort()

print(noten)  # Verschiebe das Print-Statement oder kommentiere Zeilen aus um die Veränderungen der Liste zu beobachten
```

Ausserdem gibt es eine Reihe von [*built-in* Funktionen](https://docs.python.org/3/library/functions.html), welche Informationen aus Listen abrufen oder sie anderweitig als Parameter nutzen. Meistens können diese Funktionen auch andere sequentielle Objekte als Parameter annehmen (wir werden weitere bald kennenlernen). Hier ein paar Beispiele:

## Veränderliche und Unveränderliche Objekte
Dir ist vielleicht schon aufgefallen, dass wir nun zwei Funktionen gesehen haben, die eine Liste sortieren. Ist dir schon aufgefallen, wie sie sich unterscheiden? Bevor du weiterliest, überlege dir, was die Unterschiede sind. Probiere dazu den Code anzupassen.



<details>
<summary>Wenn du bereit bist, klicke hier um die Antwort zu lesen</summary>

Bei *noten.sort()* handelt es sich um eine **in-place Manipulation**. Das bedeutet, die Funktion besitzt keinen *return-Wert*, sondern verändert direkt die Liste, von der aus die Funktion ausgeführt wird (in diesem Fall also die Liste, auf welche die Variable *noten* verweist). 

Bei *sorted(noten)* hingegen handelt es sich um eine **out-of-place Manipulation**. Diese erstellt eine neue Liste, bzw. kopiert die alte Liste, und wendet die Sortierung nur auf die neue Liste an, die alte wird gelassen, wie sie ist.
</details>


Um dieses Phänomen weiter zu verstehen, ist es wichtig etwas genauer darüber zu sprechen, was Variabeln eigentlich sind. Variabeln sind eigentlich Verweise (Pointer) auf Objekte. In manchen Fällen sind diese Objekte sogenannte *primitive* Datentypen, z.B. integer und floats. In dem Fall hat eine Variable wirklich einfach einen *Wert*. Dieser Wert kann eigentlich auch nicht verändert werden im Sinne davon wie sich eine Liste verändert, wenn man sie in-place sortiert, die Variable kann nur mit einem neuen Wert überschrieben werden. Wir nennen solche Datentypen dann *unveränderlich* (non-mutable). Dazu gehören neben primitiven Datentypen auch noch ein paar andere, die wir noch kennenlernen werden.

Die allermeisten Python-Objekte, darunter auch Listen, sind jedoch *veränderlich* (mutable). 

Am beste veranschaulichen lässt sich das an einem Beispiel:

```python
# Unveränderlich
a = 5
b = a

b = b + 1

print(a)
print(b)
```

```python
# Veränderlich
a = [1, 2]
b = a

b.append(3)

print(a)
print(b)
```

Obwohl wir nur die Variable *b* verändern in unserem zweiten Beispiel ist plötzlich auch *a* verändert worden. Das liegt daran, dass a und b schlussendlich auf das selbe Objekt verweisen, und jede Veränderung betrifft dementsprechend beide Variabeln. Also was machen, wenn wir unsere Liste zwar kopieren, aber zwei unabhängige Kopien wollen? Dazu gibt es die copy()-Funktion:

```python
a = [1, 2]
b = a.copy()

b.append(3)

print(a)
print(b)
```

## String-Formattierung
Oft müssen wir einen String auf Basis von irgendwelchen anderen Variablen anpassen. Dazu gibt es verschiedene Möglichkeiten. Die offensichtliche ist erstmal, den String einfach mit einem + mit der neuen Variable zu verbinden:

```python
text = "Es ist so toll an der Zentralbibliothek Zürich programmieren zu lernen!"
zusaetzlicher_text = "Ich liebe Python!"

gesamter_text = text + zusaetzlicher_text

print(gesamter_text)
```

Das ist aber oft nicht allzu praktisch. Wollen wir zum Beispiel etwas mittendrin einfügen, müssten wir den String jedes Mal extra slicen und wieder zusammensetzen. Praktischer sind String-Formatting-Techniken. Die erste ist die [f-string Methode](https://docs.python.org/3/reference/lexical_analysis.html#f-strings) und funktioniert folgendermassen:

```python
age = 42

# String-Concatenation
text1 = "Ich werde bald " + str(age) + " Jahre alt!"
# F-String-Formatting
text2 = f"Ich werde bald {age} Jahre alt!"

print(text2)
```

Eine andere Möglichkeit ist es, die format()-Funktion zu verwenden. Gerade wenn es sich bei dem, was Du formattieren möchtest um einen komplexen Ausdruck handelt, ist diese sauberer:

```python
age = 42

# String-Concatenation
text1 = "Ich werde bald " + str(age) + " Jahre alt!"
# Format-Methode
text3 = "Ich werde bald {0} Jahre alt!".format(age)

print(text3)
```

Toll ist an beiden Methoden, dass sie auch noch erlauben, die eingefügten Objekte leicht zu manipulieren. Zum Beispiel durch Padding:

```python
age = 42

text2 = f"Ich werde bald {age:04d} Jahre alt!"

text3 = "Ich werde bald {0:04d} Jahre alt!".format(age)

print(text2)
print(text3)
```

## Regex
Wir können Strings in Python mit Regex-Ausdrücken manipulieren mittels des [re-Moduls](https://docs.python.org/3/library/re.html). Hier gibt es nur ein kurzes Beispiel, aber im Vertiefungsmodul "NLP_Named_Entity_Recognition" wird das Modul näher erklärt.

```python
import re

# einen Text an verschiedenen Zeichen splitten
text = "Datum | Vorname - Nachname | Geburtsdatum $ Note"
split_text = re.split(r"\||-|\$", text)
print(split_text)

# einen gewissen Substring suchen
text = "Datum | Vorname - Nachname | Geburtsdatum $ Note"
matches = re.findall("((\w*?)name)", text)
print(matches)
```

# Umwandlung von Datentypen

# Falls-dann-oder-sonst: Kontrolle über den Flow
Nun können wir schon mal ein paar nette lineare Programme schreiben. Je nach dem haben wir aber nicht die volle Kontrolle darüber, was in unseren Variabeln steht, z.B. wenn wir eine Datei auslesen. Wir wollen also je nach dem, was wir haben, verschiedene Abläufe festlegen. In Python nennt sich das *Flow-Control*. Unser wichtigstes Werkzeug ist dazu *if* und *else*.

```python
number = 42
divider = 4
rest = number % divider  # der Modulo-Operator gibt die Restmenge zurück

if rest:
    print(f"The number {number} cannot cleanly be divided by {divider}!")
else:
    print(f"The number {number} can cleanly be divided by {divider}!")
```

*if* und *else* sollte recht intuitiv verständlich sein: Wir eine genannte Bedingung erfüllt, dann wird der eingeschobene Codeblock ausgeführt. Wird die Bedingung nicht erfüllt, wird der Block nach dem *else* ausgeführt. Die Einschübe macht man mit der *Tabulator*-Taste, aber sie bestehen meistens aus 4 Leerschlägen (Python akzeptiert auch 2 Leerschläge, oder Tabulator-Zeichen, aber 4 Leerschläge ist empfohlen).

Aber was ist eine Bedingung? Die Bedingung muss ein Wert sein, der sich in *True* oder *False* auflösen lässt. Damit führen wir einen neuen Datentyp ein, den Wahrheitswert (boolean). Im Codeblock oben geschieht eine implizite Umwandlung: Wenn rest gleich 0 ist, dann wird die Zahl automatisch zu False umgewandelt, falls es ein anderer Wert als 0 ist, zu True.

Wahrheitswerte werden nur selten direkt in Variablen gespeichert. Meistens entstehen sie als Folge von einer Vergleichsoperation oder einer expliziten oder impliziten Umwandlung eines anderen Datentyps. Dazu gibt es weitere Funktionen, die einen Wahrheitswert zurückgeben.

```python
# Beispiel für eine Vergleichsoperation
zahl = 5
wahrheitswert = zahl > 3
if wahrheitswert:
    print("Sie ist grösser!")

# Beispiel für eine explizite Umwandlung
text = "Hallo Welt"
wahrheitswert = bool(text)
if wahrheitswert:
    print("Der Text ist nicht leer!")

# Beispiel für eine spezielle Sorte von True/False-Return
text = "Hallo Welt"
wahrheitswert = "Welt" in text
if wahrheitswert:
    print("Welt ist Teil vom Input!")

# Beispiel für eine klassenspezifische Funktion, die einen Wahrheitswert ausgibt
text = "Hallo Welt"
wahrheitswert = text.islower()
if wahrheitswert:
    print("Der Text ist kleingeschrieben")
```

Wie hier auch zu sehen ist, ist das *else* optional.

Jedes Objekt definiert über seine `__bool__`-Methode selbst, wann es True und False zurückgibt, wenn es implizit oder explizit als Wahrheitswert verwendet wird. Zahlenwerte zum Beispiel geben `False` zurück, wenn der Zahlenwert 0 beträgt, ansonsten `True`. Strings geben nur dann `False` zurück, wenn keine Zeichen im String enthalten sind. Listen, wenn keine Elemente in der Liste enthalten sind. Diese Zustände sind nicht zu verwechseln mit `None`, denn die Objekte existieren, sie sind nur leer (Wobei ein `None`-Wert auch implizit einen Wahrheitswert von `False` zurückgibt).

Um zu zeigen, welche Bedingungen zuerst evaluiert werden sollen, müssen wir Klammern verwenden. # Eins nach dem Anderen: Schleifen
Nun folgt das letzte absolut essentielle Element, damit du deine ersten Skripte schreiben kannst! Wir haben bereits sequentielle Objekte wie Listen und Strings kennengelernt. Manchmal möchten wir uns die einzelnen Elemente in diesen Objekten genauer ansehen, mit ihnen arbeiten oder sie verändern. In dem Fall benötigen wir Schleifen (loops).

## for-Schleifen
Die üblichste Form der Schleife ist die for-Schleife. Dabei gehen wir eine Sequenz durch und definieren einen Ablauf, der für jedes Element durchgeführt wird. Dabei wird dem jeweiligen Element, an dem wir gerade sind, ein bestimmter Variablenname gegeben. Wir nennen diesen Prozess "eine Iteration". Das sieht dann ungefähr so aus:

```python
sentence = "Hier in Zürich lernen wir Python !".split()

for word in sentence:
    print(word)
```

Nochmal Schritt für Schritt. Unser Input ist also eine Liste von Wörtern. Wir *iterieren* diese Liste. Das bedeutet, für jedes Element, das in der Liste enthalten ist, wird der Ablauf, der durch die Einschübe definiert ist, durchgeführt. Während der Iteration verweist die Variable *word* auf das jeweilige Element bzw. Objekt. In der ersten Iteration also auf "Hier", in der zweiten auf "in", etc.

Wir können Schleifen natürlich auch wieder verschachteln:

```python
sentence = "Hier in Zürich lernen wir Python !".split()

for word in sentence:
    for character in word:
        print(word, character)
```

Da sowohl Listen als auch Strings Sequenzen sind, können wir beide in einer for-Loop iterieren.

## while-Schleifen
Eine weitere Art von Schleife ist die *while*-Schleife. Diese hat eine Abbruchbedingung und läuft so lange, bis diese Abbruchbedingung *True* zurückgibt. Fast immer, wenn es darum geht Sequenzen zu iterieren, ist man mit einer for-Schleife besser bedient, aber nicht alle Schleifen iterieren Sequenzen!

```python
# das Beispiel von oben, als while-Schleife
sentence = "Hier in Zürich lernen wir Python !".split()

idx = 0
while idx < len(sentence):
    word = sentence[idx]
    print(word)
    
    idx += 1
```

Sehr viel mühsamer als mit einer for-Schleife, nicht wahr? Aber in manchen Situationen kann ist die *while*-Schleife praktischer:

```python
# der User soll eine Sequenz von Zahlen eingeben, die Loop stoppt erst, wenn der User "stop" eingibt.
zahlen = []

zahl = input("Schreibe eine Zahl (or 'stop'): ")
while zahl != "stop":
    zahlen.append(zahl)
    print("You entered:", zahl)
    zahl = input("Enter a number (or 'stop'): ")

print(zahlen)
```

```python
import random

zufallszahl = random.random()
while zufallszahl < 0.9:
    print(zufallszahl)
    zufallszahl = random.random()
```

Beide Beispiele können *theoretisch* mit einer for-Schleife in Kombination mit einem Unendlichen-Generator-Objekt geschrieben werden, aber mit while lassen sich die Abläufe sehr viel natürlicher implementieren.

Vorsicht: Wenn Du while-Schleifen schreibst, achte sorgfältig darauf, dass Du Deine Abbruchbedingung irgendwann erfüllt wird. Endlosschleifen können im schlimmsten Fall Dein System abstürzen lassen, wobei Python sie eigentlich vorher stoppen sollte.

## Faule, aber praktische, Iteratoren
Ein fauler (lazy) Iterator ist ein iterierbares Objekt, das einen Wert erst generiert, wenn er gebraucht wird. Das kann zum Beispiel praktisch sein, damit man einen grossen Datensatz nicht auf einmal in den Speicher lädt, sondern immer nur die Zeile, die man gerade benötigt:

```python
# Dateien bearbeiten wird später noch erklärt
with open("./Materialien/moby_dick.txt", mode="r", encoding="utf8") as datei:
    for zeile in datei:
        print(zeile.strip())
```

Im Kontext von Schleifen, existieren zwei sehr praktische Iteratoren dieser Art, die beide durch built-in Funktionen erzeugt werden. Sieh dir die Beispiele an und versuche nachzuvollziehen, was die beiden Funktionen tun:

```python
sentence = "Hier in Zürich lernen wir Python !".split()

for index in range(0, len(sentence), 2):
    word = sentence[index]
    print(word)
```

```python
sentence = "Hier in Zürich lernen wir Python !".split()

for index, word in enumerate(sentence):
    print(index, word)
```

enumerate() nimmt eine Sequenz als Parameter, und generiert dann für jede Iteration die Zahl der Iteration. Wenn das optionale *start*-Parameter definiert wird, werden die Zahlen entsprechend modifiziert (Probier es aus!). enumerate gibt für jede Iteration eine Sequenz aus, das am ersten Index die Zahl hat, und an der zweiten Stelle das ursprüngliche Element der Liste. ## Unpacking
Unpacking ist ein sehr praktisches Feature, wenn wir mit Sequenzen von Sequenzen arbeiten. Ohne Unpacking zum Beispiel müssten wir folgendes machen:

```python
personen = [
    ["Max", 26],
    ["Lena", 13],
    ["Anna", 45],
    ["Theo", 9]
]

for person in personen:
    name = person[0]
    alter = person[1]
    print(f"{name} ist {alter} Jahre alt!")
```

Dank Unpackings geht das aber einfacher:

```python
personen = [
    ["Max", 26],
    ["Lena", 13],
    ["Anna", 45],
    ["Theo", 9]
]

for name, alter in personen:
    print(f"{name} ist {alter} Jahre alt!")
```

Was passiert aber, wenn unsere Daten unordentlich sind?

```python
personen = [
    ["Max", 26],
    ["Lena", 13],
    ["Anna", 45],
    ["Theo"]
]

for name, alter in personen:
    print(f"{name} ist {alter} Jahre alt!")
```

## Praktische Keywords: break, continue, pass

Mit *break* können wir eine Schleife vorzeitig stoppen. In einer for-Loop funktioniert also ein *break* in einer *if*-Bedingung wie die Abbruchbedingung einer *while*-Loop.

```python
sentence = "Hier in Zürich lernen wir Python !".split()

for word in sentence:
    print(word)
    if word == "Zürich":
        print("Beende Schleife, wir haben Zürich gefunden!")
        break
    
```

Wir können ausserdem ein *else* an die for-Schleife hängen, die nur dann ausgeführt wird, wenn die Schleife nicht durch ein *break* abgebrochen wurde.

```python
sentence = "Hier in Bern lernen wir Python !".split()

for word in sentence:
    print(word)
    if word == "Zürich":
        print("Beende Schleife, wir haben Zürich gefunden!")
        break
else:
    print("Wir sind wohl nicht in Zürich 🤔")
```

Ein *continue* lässt uns den Rest der derzeitigen Iteration überspringen, lässt die Schleife aber weiter laufen.

```python
sentence = "Hier in Zürich lernen wir Python !".split()

for word in sentence:
    if word == "Zürich":
        print("WO SIND WIR?")
        continue
    print(word)
```

*pass* ist lediglich ein Platzhalter, meistens wird er verwendet, wenn später dort noch Code hinkommt.

```python
sentence = "Hier in Bern lernen wir Python !".split()

for word in sentence:
    print(word)
    if word == "Zürich":
        pass
```

Das Ganze lässt sich natürlich auch kombinieren:

```python
data = [
    ["Max", 26],
    ["Lena", 13],
    ["Anna", 45],
    ["Theo", 9]
]

for name, age in data:
    if name == "Max":
        continue  # überspringen

    if name == "Lena":
        pass  # TODO

    if name == "STOP":
        break  # Beenden der Schleife!

    print(name)
else:
    print("Loop executed successfully!")
```

## Listenkomprehension

Wir können Schleifen auch sehr kompakt schreiben, und diese Schreibweise ist sogar effizienter in der Berechnung. 

Hier ein Beispiel, wie Listenkomprehension einfach benutzt werden kann, um eine neue Liste mit veränderten Elementen zu generieren:

```python
sentence = "Hier in Zürich lernen wir Python !".split()

lower_words = [word.lower() for word in sentence]

print(lower_words)
```

Die Struktur der Komprehension ist also immer [**y** for **x** in **z**]

Dabei ist **z** die Sequenz die iteriert wird, **x** ist die Variable, welche das jeweilige Element beschreibt und **y** ist ein Ausdruck, der mit diesem Element etwas tut und der neuen Liste hinzugefügt wird.

Gerne wird Listenkomprehension auch verwendet um Listen zu filtern:

```python
sentence = "Hier in Zürich lernen wir Python !".split()

upper_words = [word for word in sentence if word[0].isupper()]

print(upper_words)
```

Alternativ können wir auch ein *else* einbauen, um einen alternativen Wert oder Ausdruck in die neue Liste zu speichern.

```python
sentence = "Hier in Zürich lernen wir Python !".split()

upper_words = [word if word[0].isupper() else "_" for word in sentence]

print(upper_words)
```

# Ein Blick unter die Haube: Funktionen
Sehen wir uns, bevor wir weitermachen mit Textmanipulation, erstmal an, was eigentlich Funktionen genauer sind. Wie schon oben beschrieben, bieten sie vorgefertigte Algorithmen, nehmen Parameter entgegen und haben manchmal einen *return*-Wert. Funktionen kann man auch selbst schreiben, wenn man vermeiden möchte gewisse Codeblöcke ständig wiederholen zu müssen. Auch für ein ordentliches Skript ist das Schreiben von Funktionen empfohlen, aber als Anfänger braucht ihr das noch nicht dringend. Hier die Grundstruktur einer (recht unnötigen) Funktion:

```python
def im_quadrat(basis):
    resultat = basis*basis
    return resultat
```

Du bemerkst, wenn die Zelle ausgeführt wird, passiert erstmal gar nichts. Wir haben sie nun definiert, aber noch nicht ausgeführt. Ausführen können wir die Funktion dann so:

```python
resultat = im_quadrat(5)
print(resultat)
```

Alle Parameter, die Du bisher gesehen hast, waren *positionale* Parameter. Das bedeutet, Python guckt einfach nach der Reihenfolge, welcher Inputs in welche Parameter der Funktion gehören. Es gibt dazu optionale, positionale Parameter, die nicht unbedingt nötig sind, aber für zusätzliches Verhalten definiert werden können. Werden sie nicht definiert, erhalten sie einem *default*-Wert. *enumerate()* nimmt zum Beispiel ein solches optionales, positionales Parameter entgegen. Schliesslich können wir Parameter auch noch per keywords zuweisen. Sehen wir uns zum Beispiel sorted() an, könnten wir bestimmen, dass wir die Sortierung lieber verkehrt möchten:

```python
text = "Es ist so toll an der Zentralbibliothek Zürich programmieren zu lernen!".split()

verkehrt_sortiert = sorted(text, reverse=True)

print(verkehrt_sortiert)
```
