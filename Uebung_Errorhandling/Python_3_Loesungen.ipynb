```python
# 3.1
# Ziel: Alle Jahreszahlen im Text identifizieren.
text = "Amilcare Ponchielli ( 1834 – 1886 ) was an Italian composer . Born in Paderno Fasolaro ( now Paderno Ponchielli ) near Cremona, then Kingdom of Lombardy – Venetia, Ponchielli won a scholarship at the age of nine to study music at the Milan Conservatory , writing his first symphony by the time he was ten years old .".split()

for token in text:
    if token.isnumeric() and len(token) == 4:
        print("Jahreszahl:", token)

# NameErrors kommen auf, wenn eine Variable, Funktion oder Klasse aufgerufen wird, die nicht definiert wurde
```

```python
# 3.2
# Ziel: Aus einem String eine Liste von Brigrammen extrahieren => [("Amilcare", "Ponchielli"), ("Ponchielli", "(")]
text = "Amilcare Ponchielli ( 1834 – 1886 ) was an Italian composer . Born in Paderno Fasolaro ( now Paderno Ponchielli ) near Cremona, then Kingdom of Lombardy – Venetia, Ponchielli won a scholarship at the age of nine to study music at the Milan Conservatory , writing his first symphony by the time he was ten years old .".split()

bigrams = []
for i in range(0, len(text) - 1):
    bigram = (text[i], text[i+1])
    bigrams.append(bigram)

print(bigrams)

# In diesem Fall müssen wir schon einen Index vor dem letzten Element aufhören,
# damit das zweite Wort des letzten Bigrams noch in der Liste liegt.
# Der beste Weg um N-Grams mit Python zu extrahieren ist übrigens: https://stackoverflow.com/questions/14473477/lists-ngrams-with-zip
```

```python
# 3.3
# Ziel: Zwei Listen in einem Dictionary vereinen

scifi_authors = [
    ["Arthur", "Charles", "test"],
    ["Ursula", "Kroeber", "Le Guin"],
    ["Isaac", "Asimov"],
    ["Frank", "Herbert"]
]

scifi_works = [
    "2001: A Space Odyssey",
    "The Left Hand of Darkness",
    "I, Robot",
    "Dune"
]

scifi_archive = {}

for author, work in zip(scifi_authors, scifi_works):
    author = tuple(author)
    scifi_archive[author] = work

print(scifi_archive)

# TypeErrors kommen in vielen Formen und Farben, aber sie deuten grundsätzlich an, dass ein Objekt, das als Argument
# gegeben wurde, einen nicht zulässigen Typ hat. In diesem Fall kann man Listen nicht hashen, also müssen wir sie
# zu tuples machen.
```

```python
# 3.4
# Ziel: Welches Buch hat Clarke geschrieben?
scifi_archive = {('Arthur', 'Charles', 'Clarke'): '2001: A Space Odyssey', ('Ursula', 'Kroeber', 'Le Guin'): 'The Left Hand of Darkness', ('Isaac', 'Asimov'): 'I, Robot', ('Frank', 'Herbert'): 'Dune'}

def get_book(author):
    return scifi_archive[author]

print(get_book(("Arthur", "Charles", "Clarke")))

# KeyErrors werden aufgerufen, wenn der Schlüssel im Dictionary nicht vorkommt.
```
